```python
data = pd.read_csv('/content/drive/MyDrive/CAPSTONE WORK/data_male_female.csv')
```

```python
data.head(3)
```

```python
import gender_guesser.detector as gender
d = gender.Detector(case_sensitive=False)

name_list = []
article_id_list = []
date_list = []

for i in range(0, len(data)):
    full_names = str(data['author:string[]'].iloc[i]).split('|')
    for name in full_names:
        first_name = str(name).split(" ")[0]
        if d.get_gender(first_name) == 'male':
          name_list.append(name)
          article_id_list.append(data['article:ID'].iloc[i])
          date_list.append(data['mdate:date'].iloc[i])
```

```python
len(name_list), len(article_id_list), len(date_list)
```

```python
namedatearticle = pd.DataFrame(list(zip(name_list, article_id_list, date_list)),
               columns =['Name', 'Article_ID', 'Date'])
```
