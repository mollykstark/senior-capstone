**Generate correctly formatted json | DON'T RUN IF YOU ALREADY HAVE THE FILE**

```python
# OG represents numbers with following schema - "blank" : NumberInt(0)
# Need "blank" : 0
# this creates a new json that abides by this rule

f=open('/content/drive/MyDrive/dblpv13_clean.json', 'w')
with open('/content/drive/MyDrive/dblpv13.json','r',errors='ignore') as myFile:
  for line in myFile:
    line=line.replace("NumberInt(","").replace(")","")
    f.write(line)
f.close()
```
