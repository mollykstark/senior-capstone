```python
import gender_guesser.detector as gender
d = gender.Detector(case_sensitive=False)

primary_cont = []

for i in range(0, len(df)):
    full_names = str(df['author:string[]'].iloc[i]).split('|')
    count = 0
    for name in full_names:
        first_name = str(name).split(" ")[0]
        if count == 0 and d.get_gender(first_name) == 'female':
          primary_cont.append(True)
        if count == 0 and d.get_gender(first_name) == 'male':
          primary_cont.append(False)
        count += 1
```

```python
len(primary_cont)
```

```python
df["primary_contributor"] = primary_cont
```

```python
df_w = df[df.primary_contributor == True]
df_w = df_w[df_w['year:int'] >= 2000]
```

```python
w_counts = df_w.groupby('year:int').count()['article:ID']
```

```python
w_counts
```

```python
rates = []
for i in range(0, len(w_counts)-1):
  rates.append((w_counts.iloc[i+1]-w_counts.iloc[i])/(w_counts.iloc[i]+w_counts.iloc[i+1]))
```

```python
years = list(range(2000,2023))
```

```python
female_rates = pd.DataFrame(zip(years,rates))
```

```python
female_rates
```

```python
df_m = df[df.primary_contributor == False]
df_m = df_m[df_m['year:int'] >= 2000]
```

```python
m_counts = df_m.groupby('year:int').count()['article:ID']
```

```python
rates_m = []
for i in range(0, len(m_counts)-1):
  rates_m.append((m_counts.iloc[i+1]-m_counts.iloc[i])/(m_counts.iloc[i]+m_counts.iloc[i+1]))
```

```python
growth_rates = pd.DataFrame(zip(years,rates,rates_m), columns=["Years", "Female_Growth_Rate", "Male_Growth_Rate"])
```
